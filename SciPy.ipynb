### Advanced SciPy Learning Guide (Jupyter Notebook Format)

```python
import numpy as np
import scipy.stats as stats
import scipy.linalg as la
import scipy.optimize as opt
import scipy.integrate as integrate
import scipy.spatial.distance as dist
import matplotlib.pyplot as plt

# DARK COLORS
TEXT_COLOR = "white"
BG_COLOR = "black"

plt.rcParams["axes.facecolor"] = plt.rcParams["figure.facecolor"] = BG_COLOR
plt.rcParams["text.color"] = TEXT_COLOR
plt.rcParams["axes.labelcolor"] = TEXT_COLOR
plt.rcParams["xtick.color"] = TEXT_COLOR
plt.rcParams["ytick.color"] = TEXT_COLOR
plt.rcParams.update({
	"axes.grid" : True,
	"grid.color": "green",
	"grid.alpha": 0.35,
	"grid.linestyle": (0, (10, 10)),
})

# BETTER SIZES
DEFAULT_W, DEFAULT_H = (16, 9)
plt.rcParams["figure.figsize"] = [DEFAULT_W, DEFAULT_H]
plt.rcParams["font.size"] = 14
plt.rcParams["figure.dpi"] = 90
```

## 1. Statistical Analysis with SciPy

```python
print("--- Statistical Analysis ---")
data = np.random.normal(loc=50, scale=10, size=1000)
mean = np.mean(data)
std_dev = np.std(data)
print(f"Mean: {mean}, Standard Deviation: {std_dev}")
```

# Perform a t-test

```python
sample1 = np.random.normal(50, 10, 100)
sample2 = np.random.normal(55, 10, 100)
t_stat, p_value = stats.ttest_ind(sample1, sample2)
print(f"T-Test p-value: {p_value}")
```

## 2. Linear Algebra Operations

```python
print("\n--- Linear Algebra ---")
matrix_a = np.array([[2, 3], [4, 5]])
matrix_b = np.array([[1, 2], [3, 4]])
print("Matrix Multiplication:\n", np.matmul(matrix_a, matrix_b))
print("Determinant of A:", la.det(matrix_a))
print("Eigenvalues of A:", la.eig(matrix_a)[0])
```

## 3. Optimization with SciPy

```python
print("\n--- Optimization ---")
def objective_function(x):
    return x**2 + 3*x + 2

result = opt.minimize(objective_function, x0=0)
print("Optimized Value:", result.x)
```

## 4. Numerical Integration

```python
print("\n--- Numerical Integration ---")
def integrand(x):
    return np.sin(x)

integral, error = integrate.quad(integrand, 0, np.pi)
print(f"Integral of sin(x) from 0 to π: {integral}")
```

## 5. Distance Metrics and Clustering

```python
print("\n--- Distance Metrics ---")
points = np.array([[0, 0], [3, 4], [5, 12]])
pairwise_distances = dist.pdist(points, metric='euclidean')
print("Pairwise Euclidean Distances:", pairwise_distances)
```

## 6. Bioinformatics Example: Curve Fitting for Gene Expression

```python
print("\n--- Bioinformatics Example: Curve Fitting ---")
def sigmoid(x, a, b, c):
    return c / (1 + np.exp(-a * (x - b)))

x_data = np.linspace(0, 10, 100)
y_data = sigmoid(x_data, 1, 5, 100) + np.random.normal(0, 5, 100)
popt, _ = opt.curve_fit(sigmoid, x_data, y_data, p0=[1, 5, 100])

plt.scatter(x_data, y_data, label="Data")
plt.plot(x_data, sigmoid(x_data, *popt), color='red', label="Fitted Curve")
plt.xlabel("Gene Expression Level")
plt.ylabel("Response")
plt.title("Curve Fitting for Gene Expression Data")
plt.legend()
plt.show()
```
