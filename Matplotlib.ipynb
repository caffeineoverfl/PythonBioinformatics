### Advanced Matplotlib Learning Guide

```python
import numpy as np
import matplotlib.pyplot as plt

# DARK COLORS
TEXT_COLOR = "white"
BG_COLOR = "black"

plt.rcParams["axes.facecolor"] = plt.rcParams["figure.facecolor"] = BG_COLOR
plt.rcParams["text.color"] = TEXT_COLOR
plt.rcParams["axes.labelcolor"] = TEXT_COLOR
plt.rcParams["xtick.color"] = TEXT_COLOR
plt.rcParams["ytick.color"] = TEXT_COLOR
plt.rcParams.update({
	"axes.grid" : True,
	"grid.color": "green",
	"grid.alpha": 0.35,
	"grid.linestyle": (0, (10, 10)),
})

# BETTER SIZES
DEFAULT_W, DEFAULT_H = (16, 9)
plt.rcParams["figure.figsize"] = [DEFAULT_W, DEFAULT_H]
plt.rcParams["font.size"] = 14
plt.rcParams["figure.dpi"] = 90
```

## 1. Creating Advanced Plots

```python
print("--- Creating Advanced Plots ---")
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.figure(figsize=(8,5))
plt.plot(x, y1, label='Sine Wave', linestyle='--', color='blue')
plt.plot(x, y2, label='Cosine Wave', linestyle='-', color='red')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sine and Cosine Waves')
plt.legend()
plt.grid()
plt.show()
```

## 2. Scatter Plots with Custom Markers

```python
print("\n--- Scatter Plots ---")
x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.randint(20, 200, size=50)
colors = np.random.rand(50)
plt.scatter(x, y, s=sizes, c=colors, alpha=0.6, cmap='viridis')
plt.colorbar()
plt.title('Scatter Plot with Variable Sizes & Colors')
plt.show()
```

## 3. Histograms & Density Plots

```python
print("\n--- Histograms & Density Plots ---")
data = np.random.normal(50, 10, 1000)
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Normally Distributed Data')
plt.show()
```

## 4. Subplots & Multiple Figures

```python
print("\n--- Subplots Example ---")
# Ensure x and y have the same shape
x = np.linspace(0, 10, 100)  # 100 points
y1 = np.sin(x)  # Must match x's shape
y2 = np.cos(x)

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 6))

# Plot sine wave
axs[0, 0].plot(x, y1, 'b-')  # Ensure x and y1 have the same length
axs[0, 0].set_title('Sine Wave')

# Scatter plot (random data, not using x to avoid conflicts)
axs[0, 1].scatter(np.random.rand(50), np.random.rand(50), c='red')
axs[0, 1].set_title('Scatter Plot')

# Histogram (generated new data)
data = np.random.normal(50, 10, 1000)
axs[1, 0].hist(data, bins=20, color='green')
axs[1, 0].set_title('Histogram')

# Bar chart
axs[1, 1].bar(['A', 'B', 'C'], [5, 7, 3], color='purple')
axs[1, 1].set_title('Bar Chart')

plt.tight_layout()
plt.show()
```

## 5. 3D Plotting

```python
print("\n--- 3D Plotting ---")
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
ax.plot_surface(X, Y, Z, cmap='coolwarm')
ax.set_title('3D Surface Plot')
plt.show()
```

## 6. Bioinformatics Example: Gene Expression Visualization

```python
print("\n--- Gene Expression Visualization ---")
genes = ['BRCA1', 'TP53', 'MYC', 'EGFR', 'VEGF']
expression_levels = [23, 45, 12, 30, 18]
plt.bar(genes, expression_levels, color=['blue', 'red', 'green', 'purple', 'orange'])
plt.xlabel('Genes')
plt.ylabel('Expression Level')
plt.title('Gene Expression Levels')
plt.show()
```
